```python
import tensorflow as tf
import tensorflow.examples.tutorials.mnist.input_data as input_data
logs_path = "/tmp/mnist/"

# Load mnist data set (train, validation, test)
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)

# Config params
learning_rate = 0.1
batch_size = 100
n_hidden_1 = 16 
n_hidden_2 = 16 
training_epochs = 50
n_input = 784
n_classes = 10

# Input
with tf.name_scope("Input"):
    X = tf.placeholder(tf.float32, shape=[None, n_input], name="input_X")
    Y = tf.placeholder(tf.float32, shape=[None, n_classes], name="labels_Y")

# Weights and biases for hidden layers
with tf.name_scope("Hidden_1"):
    W1 = tf.Variable(tf.random_normal([n_input, n_hidden_1]), name="Weights")
    b1 = tf.Variable(tf.random_normal([1, n_hidden_1]), name="biases")
    A1 = tf.nn.sigmoid(tf.matmul(X, W1) + b1)

# Weights and biases for hidden layers
with tf.name_scope("Hidden_2"):
    W2 = tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2]), name="Weights")
    b2 = tf.Variable(tf.random_normal([1, n_hidden_2]), name="biases")
    A2 = tf.nn.sigmoid(tf.matmul(A1, W2) + b2)

# Output layer
with tf.name_scope("Output"):
    W3 = tf.Variable(tf.random_normal([n_hidden_2, n_classes]), name="Weights")
    b3 = tf.Variable(tf.random_normal([1, n_classes]), name="biases")
    output_layer = tf.matmul(A2, W3) + b3
    yHat = tf.nn.softmax(output_layer)

# Cost
with tf.name_scope("Cost"):
    cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=output_layer, labels=Y))
    correct_prediction = tf.equal(tf.argmax(Y, 1), tf.argmax(yHat, 1)) 
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

# Optimizer
with tf.name_scope("Optimizer"):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    train_step = optimizer.minimize(cost)

tf.summary.scalar("cost", cost)
tf.summary.scalar("accuracy", accuracy)
metrics = tf.summary.merge_all()

# Create session and train the model
sess = tf.Session()
sess.run(tf.global_variables_initializer())
writer = tf.summary.FileWriter(logs_path, graph=tf.get_default_graph())

for epoch in range(training_epochs):
    total_batches = int(mnist.train.num_examples/batch_size)
    for batch in range(total_batches):
        batch_xs, batch_ys = mnist.train.next_batch(batch_size)
        cv, _, summary = sess.run([cost, train_step, metrics], feed_dict={X:batch_xs, Y:batch_ys})
        a = sess.run([accuracy], feed_dict={X:mnist.test.images, Y:mnist.test.labels})

        # Generate summary event for TensorBoard visualization
        step = epoch * total_batches + batch
        writer.add_summary(summary, step)
        print (epoch, batch, cv, a)

```
